```python
df = pd.read_csv('../data/Table_S1.csv')


fig_1 = []
fig_4_S7 = []
for s in df.itertuples():
    if s.sample in Figure1:
        fig_1.append('yes')
    else:
        fig_1.append('no')
    if s.sample in Figure4_S7:
        fig_4_S7.append('yes')
    else:
        fig_4_S7.append('no')


df['Fig1'] = fig_1
df['Fig2'] = 'yes'

df['Fig3'] = 'G'
df.loc[df['dataset'] == 'visium_hd', 'Fig3AB'] = 'AB'
df.loc[df['group'] == 'Melanoma', 'Fig3'] = 'E,G'
df.loc[df['group'].str.startswith('PBMC'), 'Fig3'] = 'D,G'
df.loc[df['sample'].str.startswith('PBMC_F'), 'Fig3'] = 'C,D,G'
df.loc[df['group'].str.contains('OvaryCA'), 'Fig'] = 'F,G'


df['Fig4'] = fig_4_S7

df['FigS1'] = df['Fig1']


df['FigS2'] = 'no'
df.loc[((df['full_saturation'] >= 80) & (df['Fig1'] == 'yes')), 'FigS2'] = 'yes'

df['FigS3'] = df['FigS1']
df['FigS4'] = df['FigS2']


df['FigS5'] = 'no'
df.loc[df.group.isin(set(('Brain', 'Colon', 'Heart'))), 'FigS5'] = 'B'
df.loc[df.group == 'visium', 'FigS5'] = 'A'

df['FigS6'] = 'no'
df.loc[ (~df.group.str.contains('visium')) & (~df.group.str.contains('flex')), 'FigS6'] = 'yes'
df['FigS7'] = 'no'
df['FigS7'] = fig_4_S7
```

```python
df.to_csv('../Figures/Table S1.csv', index=False)
```
